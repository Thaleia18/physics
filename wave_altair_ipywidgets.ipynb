```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import altair as alt
from altair import *
from ipywidgets import widgets, interact, interactive, fixed, interact_manual
from IPython.display import display
import numpy as np
```

```python
def wavesol(A,freq,t):
    y=A*np.sin(freq*t)
    return y

def waveplot(A,freq,t):
    return plt.plot(time,wavesol(A,freq,t))
```

```python
time = np.arange(0, 20, 0.2)

interact(waveplot,A=[.1,1,10,100],freq=[.5,1,5,10,50,100],t=fixed(time))
```

```python
alt.data_transformers.disable_max_rows()

print('Altair interactive plot')
print('click in the brandcategory to highlight the point in the category')
print('mouse over a data point to see the brand name')
print('scroll over a region to zoom in.')
selector = alt.selection_single(empty='all', fields=['brandcategory'])
colours_obj = alt.Color( 'brandcategory:N')#,scale=alt.Scale(domain=parties,range=party_colours))
colours_condition = alt.condition(selector, 
                                  colours_obj,       
                                  alt.value("lightgray"))

alt.Chart(df[(df['followers']<=500000) & (df['likes']<=20000)]).mark_point().encode(
    alt.X('followers:Q',
        scale=alt.Scale(domain=(-5, 500000))),
    alt.Y('likes:Q',
          scale=alt.Scale(domain=(-5,20000))),
    size=alt.Size('comments:Q',
        scale=alt.Scale(range=(20,500))),
    color=colours_condition,
    tooltip=['brandname:N']).interactive().add_selection(selector)
```

```python
df = pd.DataFrame({'time': time})
                   
for i in [.1,1,5,10]:
    for j in [.1,1,5,10]:
        df['Amplitude=' + str(i) + ', freq=' + str(j)] = wavesol(i,j,time) # Add new column which is the result of multiplying columns i and j together

selector = alt.selection_single(empty='all', fields=['variable'])
colours_obj = alt.Color( 'variable:N')#,scale=alt.Scale(domain=parties,range=party_colours))
colours_condition = alt.condition(selector, 
                                  colours_obj,       
                                  alt.value("lightgray"))
data = df.reset_index().melt(id_vars=['time'])
alt.Chart(data[101:]).mark_line().encode(
    alt.X('time:Q',
        scale=alt.Scale(domain=(-2,22))),
    alt.Y('value:Q',
          scale=alt.Scale(domain=(-12,12))),
    color='variable'
).add_selection(selector)
```
